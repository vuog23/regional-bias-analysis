```python
import pandas as pd
import numpy as np
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
import os
from pathlib import Path
```

```python
def bytes2images(df, root_path=r"D:\Project\DAP_paper\datasets\GeoDE\processed"):
    for i in range(len(df)):
        img_bytes = df.iloc[i]["image"]
        obj = df.iloc[i]["object"]
        reg = df.iloc[i]["region"]

        img_path = img_bytes["path"]

        img = Image.open(BytesIO(img_bytes['bytes'])).convert("RGB")
        img = img.resize((256, 256), Image.Resampling.LANCZOS)

        save_dir = Path(root_path) / reg / obj
        save_dir.mkdir(parents=True, exist_ok=True)

        filename = Path(img_path).name

        save_path = save_dir / filename

        img.save(save_path)
```

```python
parquet_folder = r"D:\Project\DAP_paper\datasets\GeoDE\raw"
parquets = os.listdir(r"D:\Project\DAP_paper\datasets\GeoDE\raw")

for parquet in parquets:
    df_path = os.path.join(parquet_folder, parquet)
    df = pd.read_parquet(df_path)
    bytes2images(df)
```
